## Task 1

In Task 1, the Python Function replicates the NDOT mix design calculations. It calculates the important mix design numbers for one cubic yard of concrete so later calculation can work correctly.


```python
cubic_yard_ft3 = 27 
unit_water_weight = 62.4

print(f"Define constant: 1 cubic yard = {cubic_yard_ft3} cubic feet")
print(f"Unit weight of water = {unit_water_weight} lb per cubic feet")
```

```python
# define a function to calculate water weight

def calculate_water_weight_Q(cement_A, Fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E):
    
    # Our first local variable: total cementitious weight
    total_cementitious = cement_A + Fly_ash_B + silica_fume_C + other_SCM_D 
    
    # Our second local variable: water weight
    water_weight_Q = total_cementitious * wc_ratio_E

    # this returns water weight
    return water_weight_Q
```

```python
# Define a function to calculate the volume of cement (R)

def calculate_volume_R(weight_cement_A, sg_cement_J):
    volume_R = cement_A / (sg_cement_J * unit_water_weight)
    return volume_R
```

```python
# Define a function to calculate fly ash volume

def calculate_volume_S(fly_ash_B, sg_fly_ash_K):
    volume_S = fly_ash_B / (sg_fly_ash_K * unit_water_weight)
    return volume_S
```

```python
# Define a function to calculate silica fume volume

def calculate_volume_T(silica_fume_C, sg_silica_fume_L):
    volume_T = silica_fume_C / (sg_silica_fume_L * unit_water_weight)
    return volume_T
```

```python
# Define a function to calculate other SCM volume

def calculate_volume_U(other_scm_D, sg_other_scm_M):
    volume_U = other_scm_D / (sg_other_scm_M * unit_water_weight)
    return volume_U
```

```python
# Define a function to calculate other air volume

def calculate_air_volume_V(air_content_F):
    volume_V = (air_content_F / 100 ) * cubic_yard_ft3
    return volume_V
```

```python
# Define a function to calculate water volume (W) as a function of water weight (Q)

def calculate_water_volume_W(water_weight_Q):
    volume_W = water_weight_Q / unit_water_weight
    return volume_W
```

```python
# Define a function to calculate total aggregate volume (X)

def calculate_total_aggregate_volume_X(volume_R, volume_S, volume_T, volume_U, volume_V, volume_W):
    volume_X = cubic_yard_ft3 - volume_R - volume_S - volume_T - volume_U - volume_V - volume_W
    return volume_X
```

```python
# Define a function to calculate fine aggregate weight (Y)

def calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, volume_X):
    weight_Y = unit_water_weight * (percent_fine_G / 100) * sg_fine_N * volume_X
    return weight_Y
```

```python
# Function for coarse aggregate weight (Z)

def calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, volume_X):
    weight_Z = unit_water_weight * (percent_coarse_H / 100) * sg_coarse_O * volume_X
    return weight_Z
```

```python
# Function for other aggregate weight (AA)

def calculate_other_aggregate_AA(percent_other_I, sg_coarse_P, volume_X):
    weight_AA = unit_water_weight * (percent_other_I / 100) * sg_coarse_P * volume_X
    return weight_AA
```

## Task 2

In task 2, the program collects the mix designs inputs into a step-by-step order. These inputs mirror and follow the NDOT Excel worksheet to ensure the calculations are performed correctly.

```python
# General info

project_no = int(input("Enter project number: "))
concrete_class = input("Enter class of concrete: ")

# Cementitious material inputs

cement_A = float(input("Enter cement weight A (lb per cubic yard): "))
fly_ash_B = float(input("Enter fly ash weight B (lb per cubic yard): "))
silica_fume_C = float(input("Enter silica fume weight C (lb per cubic yard): "))
other_scm_D = float(input("Enter other SCM weight D (lb per cubic yard): "))

# Design parameters

wc_ratio_E = float(input("Enter target water-cement ratio E: "))
air_content_F = float(input("Enter target air content F (%): "))

# Aggregate proportions

percent_fine_G = float(input("Enter percent fine aggregate G (%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H (%): "))
percent_other_I = float(input("Enter percent other aggregate I (%): "))

# Specific gravities

sg_cement_J = float(input("Enter specific gravity of cement J: "))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K: "))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L: "))
sg_other_scm_M = float(input("Enter specific gravity of other SCM M: "))

sg_fine_N = float(input("Enter specific gravity of fine aggregate N: "))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O: "))
sg_other_P = float(input("Enter specific gravity of other aggregate P: "))

# A Print message to confirm all inputs have been collected

print("\nAll user inputs collected successfully.")
```

## Step 3

In Task 3, the program uses the calculated values to generate a final concrete mix design ouput. These results are displayed in a clear labeled format that summarizes the mix design for one cubic yard of concrete.

```python
# Step 1: Water weight (Q)

Q = calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, wc_ratio_E)

# Step 2: Cementitious volumes (R, S, T, U)

R = calculate_volume_R(cement_A, sg_cement_J)
S = calculate_volume_S(fly_ash_B, sg_fly_ash_K)
T = calculate_volume_T(silica_fume_C, sg_silica_fume_L)
U = calculate_volume_U(other_scm_D, sg_other_scm_M)

# Step 3: Air and water volumes (V, W)

V = calculate_air_volume_V(air_content_F)
W = calculate_water_volume_W(Q)

# Step 4: Total aggregate volume (X)

X = calculate_total_aggregate_volume_X(R, S, T, U, V, W)

# Step 5: Aggregate weights (Y, Z, AA)

Y = calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, X)
Z = calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, X)
AA = calculate_other_aggregate_AA(percent_other_I, sg_other_P, X)
```

### Scenerios 1-4

```python
WATER_UNIT_WEIGHT = 62.4  # lb/ft^3
CY_VOLUME = 27.0          # ft^3 per cubic yard

def absolute_volume(weight_lb, specific_gravity):
    return weight_lb / (specific_gravity * WATER_UNIT_WEIGHT)

def run_scenario(scenario):
    print("---------------------------------------------")
    print(f"Scenario {scenario['id']}")
    print("---------------------------------------------")
    print(f"Source: {scenario['source']}")
    print(f"Use Case: {scenario['use_case']}")
    print("---------------------------------------------")

    # The inputs
    cement = scenario["cement_content_lb_yd^3"]
    wc_ratio = scenario["wc_ratio"]
    air_percent = scenario["air_content_percent"]
    sg_cement = scenario["sg_cement"]
    sg_fine = scenario["sg_fine"]
    sg_coarse = scenario["sg_coarse"]
    fine_fraction = scenario["fine_agg_fraction"]
    
    # The Calculations 
    water = cement * wc_ratio

    Vol_cement = absolute_volume(cement, sg_cement)
    Vol_water = water / WATER_UNIT_WEIGHT
    Vol_air = (air_percent / 100.0) * CY_VOLUME

    Vol_agg_total = CY_VOLUME - (Vol_cement + Vol_water + Vol_air)

    Vol_fine = Vol_agg_total * fine_fraction
    Vol_coarse = Vol_agg_total * sg_coarse * WATER_UNIT_WEIGHT

    fine_weight = Vol_fine * sg_fine * WATER_UNIT_WEIGHT
    coarse_weight = Vol_coarse * sg_coarse * WATER_UNIT_WEIGHT

    total_weight = cement + water + fine_weight + coarse_weight

    print("\nComponent\t\tWeight (lb/yd^3)\tNotes")
    print("---------------------------------------------")
    print(f"Cement\t\t\t{cement:8.1f}\tGiven input")
    print(f"Water\t\t\t{water:8.1f}\tw/c = {wc_ratio}")
    print(f"Fine Aggregate\t\t{fine_weight:8.1f}\tSG = {sg_fine}")
    print(f"Coarse Aggregate\t{coarse_weight:8.1f}\tSG = {sg_coarse}")
    print(f"Air (volumetric)\t{'-':>8}\t{air_percent}% ({Vol_air:.2f} ft^3)")
    print("---------------------------------------------")
    print(f"Total (sanity check)\t{total_weight:8.1f}\tExcludes air")

    print("\nClient Summary:")
    print(
        f"This mix design provides {cement:.0f} lb/yd^3 cement",
        f"with a {wc_ratio:.2f} water-cement ratio and {air_percent:.1f}% air content.",
        f"The total batch weight is approximately {total_weight:.0f} lb per cubic yard.",
        f"Designed for {scenario['use_case']} per cited source.",
    )

    print("\nRounding Rules Applied:")
    print("1) Weights rounded to 1 decimal place")
    print("2) Air volume shown to 2 decimals (ft^3)")
    print("3) Units clearly shown as lb/yd^3 and ft^3")

scenarios = [
    {
        "id": 1,
        "source": "NDOT Standard Specifications (Section 501  - PCC)",
        "use_case": "Highway pavement",
        "cement_content_lb_yd^3": 611,
        "wc_ratio": 0.42,
        "air_content_percent": 6.0,
        "sg_cement": 3.15,
        "sg_fine": 2.65,
        "sg_coarse": 2.72,
        "fine_agg_fraction": 0.38
    },

    {
        "id": 2,
        "source": "ACI 211.1 - Bridge Deck Concrete Guidance",
        "use_case": "Bridge deck exposure",
        "cement_content_lb_yd^3": 658,
        "wc_ratio": 0.40,
        "air_content_percent": 7.0,
        "sg_cement": 3.15,
        "sg_fine": 2.63,
        "sg_coarse": 2.68,
        "fine_agg_fraction": 0.40
    },

    {
        "id": 3,
        "source": "DOT Cold Weather Concreting Specification",
        "use_case": "Cold Weather structural placement",
        "cement_content_lb_yd^3": 705,
        "wc_ratio": 0.38,
        "air_content_percent": 6.5,
        "sg_cement": 3.15,
        "sg_fine": 2.66,
        "sg_coarse": 2.72,
        "fine_agg_fraction": 0.37
    },

    {
        "id": 4,
        "source": "FAA / DOT Airfield Pavement Mix Reference",
        "use_case": "Airfield pavement",
        "cement_content_lb_yd^3": 658,
        "wc_ratio": 0.41,
        "air_content_percent": 5.5,
        "sg_cement": 3.15,
        "sg_fine": 2.64,
        "sg_coarse": 2.69,
        "fine_agg_fraction": 0.36
    }
]

for scenario in scenarios:
    run_scenario(scenario)
```
